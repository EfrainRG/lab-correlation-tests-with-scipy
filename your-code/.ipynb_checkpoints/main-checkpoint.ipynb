# Challenge 1 - Analysis of Variance

In this part of the lesson, we will perform an analysis of variance to determine whether the factors in our model create a significant difference in the group means. We will be examining a dataset of FIFA players. We'll start by loading the data using the code in the cell below.

```python
# Run this code:
fifa = pd.read_csv('fifa.csv')
```

Let's examine the dataset by looking at the `head`.

```python
# Your code here:
fifa.head()
```

Player's values are expressed in millions of euros. We would like this column to be numeric. Therefore, let's create a numeric value column. Do this by stripping all non-numeric characters from each cell. Assign this new data to `ValueNumeric`. There is no need to multiply the value to be expressed in millions. 

```python
# Your code here:
fifa['ValueNumeric'] = fifa['Value'].str.extract('(\d+)').astype(float)
fifa.dtypes
```

#### We'd like to determine whether a player's preffered foot and position have an impact on their value. 

Using the `statsmodels` library, we are able to produce an ANOVA table without munging our data. Create an ANOVA table with value as a function of position and preferred foot. Recall that pivoting is performed by the `C` function.

Hint: For columns that have a space in their name, it is best to refer to the column using the dataframe (For example: for column `A`, we will use `df['A']`).

```python
# usando scipy para solución
# Aquí no tuve idea de porque con stats.f_oneway() obtuve nan values
```

```python
fifa['foot_count'] = fifa.groupby('Preferred Foot').cumcount()
foot_pivot = fifa.pivot(index='foot_count',columns='Preferred Foot', values='ValueNumeric')
foot_pivot.columns = ['Foot_'+str(x) for x in foot_pivot.columns.values]
foot_pivot.head()
```

```python
# Your code here:
stats.f_oneway(foot_pivot.Foot_nan ,foot_pivot.Foot_Left, foot_pivot.Foot_Right) 
```

```python
fifa['position_count'] = fifa.groupby('Position').cumcount()
position_pivot = fifa.pivot(index='position_count',columns='Position', values='ValueNumeric')
position_pivot.head()
```

```python
stats.f_oneway(position_pivot.CAM, position_pivot.CB, position_pivot.CDM, 
              position_pivot.CF, position_pivot.CM, position_pivot.GK,
              position_pivot.LAM, position_pivot.LB, position_pivot.LCB,
              position_pivot.LCM, position_pivot.LDM, position_pivot.LF,
              position_pivot.LM, position_pivot.LS, position_pivot.LW, 
              position_pivot.LWB, position_pivot.RAM, position_pivot.RB, 
              position_pivot.RCB, position_pivot.RCM, position_pivot.RDM, 
              position_pivot.RF, position_pivot.RM, position_pivot.RS, 
              position_pivot.RW, position_pivot.RWB, position_pivot.ST)
```

```python
# método 'statsmodels'
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
# primero cambiaremos el nomre de la columna de 'Preferred Foot'
fifa = fifa.rename(columns={'Preferred Foot':'Preferred_Foot'})
fifa.head()
```

```python
# haciendo un análisis respecto al pie preferido y el valor del jugador
model = ols('ValueNumeric ~ C(Preferred_Foot)', data=fifa).fit()
anova_table = sm.stats.anova_lm(model, typ=1)
anova_table
```

```python
# haciendo un análisis respecto a la posición y el valor del jugador
model = ols('ValueNumeric ~ C(Position)', data=fifa).fit()
anova_table = sm.stats.anova_lm(model)
anova_table
```

```python
model = ols('ValueNumeric ~ C(Position)+C(Preferred_Foot)', data=fifa).fit()
anova_table = sm.stats.anova_lm(model)
anova_table
```

```python
# haciendo un análisis respecto a la posición junto con el pie preferido, al valor del jugador
model = ols('ValueNumeric ~ C(Position)*C(Preferred_Foot)', data=fifa).fit()
anova_table = sm.stats.anova_lm(model)
anova_table
```

```python
sns.pairplot(fifa[['Position','ValueNumeric','Preferred_Foot']],kind='reg');
```

```python
model.summary()
```

What is your conclusion from this ANOVA?

```python
# Your conclusions here:
# Despues de mirar los resultados, se pude concluir que aunque es muy poca
# la diferencia (tomandeo en cuanta un 0.95, obteniendo 8.274739e-02) por la
# cual no esta influyendo la posisción y el pie preferido respecto al valor
# del jugador
```

After looking at a model of both preffered foot and position, we decide to create an ANOVA table for nationality. Create an ANOVA table for numeric value as a function of nationality.

```python
# Your code here:
# haciendo un análisis respecto a la nacionalidad, al valor del jugador
model = ols('ValueNumeric ~ C(Nationality)', data=fifa).fit()
anova_table = sm.stats.anova_lm(model)
anova_table
```

What is your conclusion from this ANOVA?

```python
# Aquí definitivamente podemos notar que no hay relación entre la nacionalidad
# del jugador y su valor. POdemos encontrar jugadores buenos en cualquier lugar
```

# Challenge 2 - Linear Regression

Our goal with using linear regression is to create a mathematical model that will enable us to predict the outcome of one variable using one or more additional independent variables.

We'll start by ensuring there are no missing values. Examine all variables for all missing values. If there are missing values in a row, remove the entire row.

```python
fifa2 = pd.read_csv('fifa.csv')
fifa2['ValueNumeric'] = fifa2['Value'].str.extract('(\d+)').astype(float)
fifa2.dtypes
```

```python
# Your code here:
fifa2.isna().sum()
```

```python
fifa2.head(50)
```

```python
fifa2 = fifa2.dropna(axis=0)
```

```python
fifa2.isna().sum()
```

```python
fifa2.head(1)
```

Using the FIFA dataset, in the cell below, create a linear model predicting value using stamina and sprint speed. create the model using `statsmodels`. Print the model summary.

Hint: remember to add an intercept to the model using the `add_constant` function.

```python
# Using Scipy method-----------------------------------------------
```

```python
# Your code here:
from scipy.stats import linregress
```

```python
slope, intercept, r_value, p_value, std_err = linregress(fifa2.Stamina,fifa2.ValueNumeric)
slope, intercept, r_value, p_value, std_err
```

```python
slope, intercept, r_value, p_value, std_err = linregress(fifa2.SprintSpeed, fifa2.ValueNumeric)
slope, intercept, r_value, p_value, std_err
```

```python
# formula ocupando Stamina
# Value = 289.594 - 0.031 * Staminna
# como dato extra, con este modelo solo abarcaremos alrededor del 1.8%
# formula ocupando SprintSpeed
# Value = 270.220 - 0.031 * SprintSpeed
# como dato extra, con este modelo solo abarcaremos alrededor del 0.1%
```

```python
# Using statsmodels method-----------------------------------------------
```

```python
import statsmodels.api as sm
```

```python
# Stamina -----------------------
X = sm.add_constant(fifa2.Stamina)
# 'Y' Variable dependiente o la deseada a predecir
Y = fifa2.ValueNumeric
modelo_stamina = sm.OLS(Y, X).fit()
```

```python
# pidiendo las predicciones
prediccion_by_stamina = modelo_stamina.predict(X)
fifa2['predecido_by_stamina'] = prediccion_by_stamina
# para saber que funciona el modelo
modelo_impreso_stamina = modelo_stamina.summary()
modelo_impreso_stamina
```

```python
# SprintSpeed -----------------------
X = sm.add_constant(fifa2.SprintSpeed)
# 'Y' Variable dependiente o la deseada a predecir
Y = fifa2.ValueNumeric
modelo_sprintspeed = sm.OLS(Y, X).fit()
```

```python
# pidiendo las predicciones
prediccion_by_sprintspeed = modelo_sprintspeed.predict(X)
fifa2['predecido_by_sprintspeed'] = prediccion_by_sprintspeed
# para saber que funciona el modelo
modelo_impreso_sprintspeed = modelo_sprintspeed.summary()
modelo_impreso_sprintspeed
```

```python
# Stamina & SprintSpeed -----------------------
X = sm.add_constant(fifa2[['Stamina','SprintSpeed']])
# 'Y' Variable dependiente o la deseada a predecir
Y = fifa2.ValueNumeric
modelo_stamina_speed = sm.OLS(Y, X).fit()
```

```python
# pidiendo las predicciones
prediccion_by_stamina_speed = modelo_stamina_speed.predict(X)
fifa2['predecido_by_stamina_speed'] = prediccion_by_stamina_speed
# para saber que funciona el modelo
modelo_impreso_stamina_speed = modelo_stamina_speed.summary()
modelo_impreso_stamina_speed
```

```python
fifa2[['ValueNumeric','predecido_by_stamina_speed','predecido_by_stamina','predecido_by_sprintspeed']]
```

Report your findings from the model summary. In particular, report about the model as a whole using the F-test and how much variation is predicted by the model using the r squared.

```python
# Your conclusions here:
# Como podemos notar en los resultados obtenidos, los datos como 'Stamina' & 'SprintSpeed'
# no son datos o información relevantes para el valor de un jugador.
# En otras palabras podría haber jugadores que tengan 'Stamina' o 'Speed' con 
# gran número y aún así el valor númerico del jugador puede ser bajo.
# En cuanto a la variación de los datos con este modelo se obtiene menos del 1%
```

Next, create a second regression model predicting value using potential. Create the model using `statsmodels` and print the model summary. Remember to add a constant term.

```python
# Your code here:
# Stamina, SprintSpeed & Potencial -----------------------
X = sm.add_constant(fifa2[['Stamina','SprintSpeed','Potential']])
# 'Y' Variable dependiente o la deseada a predecir
Y = fifa2.ValueNumeric
modelo_stamina_speed_potential = sm.OLS(Y, X).fit()
```

```python
# pidiendo las predicciones
prediccion_by_stamina_speed_potential = modelo_stamina_speed_potential.predict(X)
fifa2['predecido_by_stamina_speed_potential'] = prediccion_by_stamina_speed_potential
# para saber que funciona el modelo
modelo_impreso_stamina_speed_potential = modelo_stamina_speed_potential.summary()
modelo_impreso_stamina_speed_potential
```

```python
# Potencial -----------------------
X = sm.add_constant(fifa2.Potential)
# 'Y' Variable dependiente o la deseada a predecir
Y = fifa2.ValueNumeric
modelo_potential = sm.OLS(Y, X).fit()
```

```python
# pidiendo las predicciones
prediccion_by_potential = modelo_potential.predict(X)
fifa2['predecido_by_potential'] = prediccion_by_potential
# para saber que funciona el modelo
modelo_impreso_potential = modelo_potential.summary()
modelo_impreso_potential
```

```python
sns.pairplot(fifa2[['Potential','ValueNumeric']],kind='reg');
```

```python
fifa2[['ValueNumeric','predecido_by_stamina_speed_potential','predecido_by_potential']]
```
